```python
import requests
from bs4 import BeautifulSoup

spark_url = "http://node0655.palmetto.clemson.edu:3727/api/v1/applications"
html_text = requests.get(spark_url).text
soup = BeautifulSoup(html_text, 'html.parser')
```

```python
print(soup)
```

```python
for link in soup.find_all('li'):
    print(link.text)
```

```python
for div in soup.find_all('div', class_='aggregated-workers collapsible-table'):
    print(div)
```

```python
th_headers = soup.find_all('th')
td_elements = soup.find_all('td')
```

## Worker Information

```python
workers = 2
for w in range(workers):
    for i in range(6):
        print(th_headers[i].text + ": " + td_elements[w * 6 + i].text.strip())
```

## Application Information

```python
headers_index = 6

for i in range(headers_index, len(th_headers)):
    print(th_headers[i].text)
for i in range(headers_index, len(th_headers)):
    print(td_elements[i].text.strip())
```
